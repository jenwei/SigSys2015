To show the effects of the follow controls, two feedback gains and their step responses are depicted below to visualize whether its possible to stabilize the system or not. The K values used are 0.5 and 5.

B. Proportional Control

The transfer function is $$\frac{K}{s^2-0.01s+1+K}$$

```python
K = 0.50
sys = signal.lti([K,0], [1, -0.01, 1+K])

t = np.linspace(0, 100, 1000)
t, s = signal.step(sys, T=t)
plt.suptitle('Step Response K=0.50')
timeplot(t, s)
```

```python
pzmap(sys)
```

```python
K = 5
sys = signal.lti([K,0], [1, -0.01, 1+K])

t = np.linspace(0, 100, 1000)
t, s = signal.step(sys, T=t)
plt.suptitle('Step Response, K=5')
timeplot(t, s)
```

```python
pzmap(sys)
```

###### As shown in the plots above, for both cases of K, the step responses oscillate and diverge, and the system is unstable (this is verified by the poles in the pole-zero maps which both lie to the right of the imaginary axis). As K increases, the oscillations densify and increase in amplitude.

C. Integral Control

The transfer function is $$\frac{K}{s^3-0.01s^2+s+K}$$

```python
K = 0.50
sys = signal.lti([K,0], [1, -0.01,  1, K])

t = np.linspace(0, 100, 1000)
t, s = signal.step(sys, T=t)
plt.suptitle('Step Response K=0.50')
timeplot(t, s)
```

```python
pzmap(sys)
```

```python
K = 5
sys = signal.lti([K,0], [1, -0.01, 1, K])

t = np.linspace(0, 100, 1000)
t, s = signal.step(sys, T=t)
plt.suptitle('Step Response K=5')
timeplot(t, s)
```

```python
pzmap(sys)
```

###### As shown in the plots above, the increase in K increases the values of the poles in both imaginary and real terms (the poles move further away from the origin). Both pole-zero maps show poles to the right of the imaginary axis, so the system will not stabilize.

D. Differential Control

The transfer function is $$\frac{K}{s^2-0.01s+Ks+1}$$

```python
K = 0.50
sys = signal.lti([K,0], [1, -0.01+K, 1])

t = np.linspace(0, 100, 1000)
t, s = signal.step(sys, T=t)
plt.suptitle('Step Response K=0.50')
timeplot(t, s)
```

```python
pzmap(sys)
```

```python
K = 5
sys = signal.lti([K,0], [1, -0.01+K, 1])

t = np.linspace(0, 100, 1000)
t, s = signal.step(sys, T=t)
plt.suptitle('Step Response K=5')
timeplot(t, s)
```

```python
pzmap(sys)
```

###### As shown in the plots above, when K=0.5, the step response oscillates, but in general, both K=0.5 and K=5 converge to 0 and stabilizes. All poles lie to the left of the imaginary axis.
